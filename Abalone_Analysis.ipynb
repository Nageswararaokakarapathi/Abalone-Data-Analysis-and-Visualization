```python
# ==========================================================
# STEP 7: CALCULATE BASIC STATISTICAL MEASURES
# ==========================================================
#
# In this step, we calculate important statistical
# measures for all numerical columns.
#
# These statistics help us understand the central
# tendency and variability of the data.
#
# The parameter numeric_only=True ensures that
# calculations are performed only on numerical columns.
#
# ==========================================================


# ----------------------------------------------------------
# 1. Mean (Average)
# ----------------------------------------------------------
#
# The mean is the average value of a column.
#
# Formula:
# Mean = Sum of all values / Total number of values
#
# It represents the central value of the data.
# ----------------------------------------------------------

print("========== MEAN ==========\n")
print(df.mean(numeric_only=True))


# ----------------------------------------------------------
# 2. Median
# ----------------------------------------------------------
#
# The median is the middle value when the data
# is arranged in ascending order.
#
# If the number of observations is even,
# the median is the average of the two middle values.
#
# Median is less affected by outliers than the mean.
# ----------------------------------------------------------

print("\n========== MEDIAN ==========\n")
print(df.median(numeric_only=True))


# ----------------------------------------------------------
# 3. Minimum Value
# ----------------------------------------------------------
#
# The minimum value represents the smallest
# observation in each numerical column.
# ----------------------------------------------------------

print("\n========== MINIMUM VALUES ==========\n")
print(df.min(numeric_only=True))


# ----------------------------------------------------------
# 4. Maximum Value
# ----------------------------------------------------------
#
# The maximum value represents the largest
# observation in each numerical column.
# ----------------------------------------------------------

print("\n========== MAXIMUM VALUES ==========\n")
print(df.max(numeric_only=True))


# ----------------------------------------------------------
# 5. Variance
# ----------------------------------------------------------
#
# Variance measures how much the data values
# are spread around the mean.
#
# A larger variance indicates greater variability,
# while a smaller variance means the values are
# closer to the mean.
# ----------------------------------------------------------

print("\n========== VARIANCE ==========\n")
print(df.var(numeric_only=True))


# ----------------------------------------------------------
# 6. Standard Deviation
# ----------------------------------------------------------
#
# Standard Deviation is the square root of variance.
#
# It measures the average distance of data points
# from the mean.
#
# Lower Standard Deviation:
# Data points are close to the mean.
#
# Higher Standard Deviation:
# Data points are more spread out.
# ----------------------------------------------------------

print("\n========== STANDARD DEVIATION ==========\n")
print(df.std(numeric_only=True))
```

Interpretation

The statistical measures provide detailed information about the numerical variables in the dataset.

Mean
Represents the average value of each feature.
Helps understand the central tendency of the data.
Median
Represents the middle value of the data.
Less affected by extreme values (outliers) than the mean.
Minimum and Maximum
Show the range of values for each variable.
Help identify the smallest and largest observations.
Variance
Measures how much the values are spread around the mean.
A higher variance indicates greater variability.
Standard Deviation
Indicates the average deviation of values from the mean.
A smaller value means the data is more consistent.
A larger value indicates greater dispersion.

These statistics are useful for understanding the overall characteristics of the dataset before visualization and modeling.

```python
# ==========================================================
# STEP 9: CALCULATE THE COVARIANCE MATRIX
# ==========================================================
#
# Covariance measures how two numerical variables
# change together.
#
# It helps us understand whether two variables
# increase or decrease together.
#
# Interpretation:
#
# Positive Covariance (+)
# -----------------------
# If one variable increases, the other variable
# also tends to increase.
#
# Negative Covariance (-)
# -----------------------
# If one variable increases, the other variable
# tends to decrease.
#
# Covariance Near Zero
# --------------------
# Indicates little or no linear relationship
# between the variables.
#
# The cov() function calculates the covariance
# between all numerical variables in the dataset.
#
# numeric_only=True ensures that only numerical
# columns are included in the calculation.
#
# ==========================================================

# Calculate the Covariance Matrix
cov_matrix = df.cov(numeric_only=True)

# Display the Covariance Matrix
print("========== COVARIANCE MATRIX ==========\n")

cov_matrix
```

```python
# ==========================================================
# STEP 9.1: VISUALIZE THE COVARIANCE MATRIX
# ==========================================================
#
# A heatmap provides a graphical representation
# of the covariance matrix.
#
# Darker colors indicate larger covariance values.
#
# Positive values show variables moving in
# the same direction.
#
# Negative values show variables moving in
# opposite directions.
#
# ==========================================================

plt.figure(figsize=(10,8))

sns.heatmap(
    cov_matrix,
    annot=True,
    cmap="Blues",
    fmt=".2f"
)

plt.title("Covariance Matrix of Abalone Dataset")

plt.show()
```

Interpretation

The covariance matrix shows the relationship between every pair of numerical variables in the dataset.

A positive covariance means that the two variables tend to increase or decrease together.
A negative covariance means that when one variable increases, the other tends to decrease.
A covariance value close to zero indicates little or no linear relationship between the variables.

From the Abalone dataset, we observe that variables such as Length, Diameter, and Whole Weight have positive covariance values, indicating that larger abalones generally have greater physical measurements and weight.

The covariance matrix is useful for understanding relationships among variables and serves as the foundation for many multivariate statistical techniques.

```python
# ==========================================================
# STEP 18: CALCULATE THE CORRELATION MATRIX
# ==========================================================
#
# Correlation measures the strength and direction
# of the relationship between two numerical variables.
#
# The correlation coefficient always lies between:
#
# +1  -> Perfect Positive Correlation
#  0  -> No Correlation
# -1  -> Perfect Negative Correlation
#
# Interpretation:
#
# +1.0  : Perfect Positive Relationship
# +0.7 to +0.9 : Strong Positive Relationship
# +0.4 to +0.6 : Moderate Positive Relationship
# +0.1 to +0.3 : Weak Positive Relationship
#
# 0     : No Relationship
#
# -0.1 to -0.3 : Weak Negative Relationship
# -0.4 to -0.6 : Moderate Negative Relationship
# -0.7 to -0.9 : Strong Negative Relationship
# -1.0 : Perfect Negative Relationship
#
# numeric_only=True ensures that only numerical
# columns are used for calculating correlations.
#
# ==========================================================

# Calculate Correlation Matrix
corr = df.corr(numeric_only=True)

# Display the Correlation Matrix
print("========== CORRELATION MATRIX ==========\n")

corr
```

```python
# ==========================================================
# STEP 18.1: VISUALIZE THE CORRELATION MATRIX
# ==========================================================
#
# A heatmap is used to visualize the correlation
# between numerical variables.
#
# Dark Red   -> Strong Positive Correlation
# Dark Blue  -> Strong Negative Correlation
#
# Values close to +1 indicate strong positive
# relationships.
#
# Values close to -1 indicate strong negative
# relationships.
#
# Values close to 0 indicate weak or no relationship.
#
# ==========================================================

import os

# Create images folder if it doesn't exist
os.makedirs("images", exist_ok=True)

# Set figure size
plt.figure(figsize=(10,8))

# Create Heatmap
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)

# Add title
plt.title("Correlation Matrix Heatmap")

# Save graph
plt.savefig(
    "images/Correlation_Heatmap.png",
    dpi=300,
    bbox_inches="tight"
)

# Display graph
plt.show()

print("✅ Correlation Heatmap saved successfully!")
```

Interpretation

The correlation matrix measures the strength and direction of the relationship between all numerical variables.

From the Abalone dataset, we can observe:

Length and Diameter have a strong positive correlation, indicating that larger abalones tend to have larger diameters.
Whole Weight, Shucked Weight, Viscera Weight, and Shell Weight are also strongly positively correlated with Length.
Rings generally show a moderate positive correlation with physical measurements, suggesting that older abalones tend to be larger.

```python
# ==========================================================
# STEP 25: BUILDING A LINEAR REGRESSION MODEL
# ==========================================================
#
# Machine Learning is used to predict values based on
# the relationship between input features and the target.
#
# Here:
#
# Target Variable (y)  -> Rings
#
# Input Features (X) ->
# Length
# Diameter
# Height
# Whole_weight
# Shucked_weight
# Viscera_weight
# Shell_weight
#
# We remove the 'Sex' column because it is categorical.
# (If categorical variables exist, we usually convert them
# using one-hot encoding.)
#
# ==========================================================

# Import required libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# ----------------------------------------------------------
# Step 1: Select Input Features (Independent Variables)
# ----------------------------------------------------------

# Drop the target variable (Rings)
# Drop Sex because it is categorical
X = df.drop(["Rings", "Sex"], axis=1)

# Convert categorical columns into numbers (safe practice)
X = pd.get_dummies(X)

# ----------------------------------------------------------
# Step 2: Select Target Variable
# ----------------------------------------------------------

# Rings is the value we want to predict
y = df["Rings"]

# ----------------------------------------------------------
# Step 3: Split Dataset
# ----------------------------------------------------------

# 80% Training Data
# 20% Testing Data

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

# ----------------------------------------------------------
# Step 4: Create Linear Regression Model
# ----------------------------------------------------------

model = LinearRegression()

# ----------------------------------------------------------
# Step 5: Train the Model
# ----------------------------------------------------------

model.fit(X_train, y_train)

# ----------------------------------------------------------
# Step 6: Predict on Test Data
# ----------------------------------------------------------

pred = model.predict(X_test)

# ----------------------------------------------------------
# Step 7: Evaluate Model Performance
# ----------------------------------------------------------

print("R² Score :", r2_score(y_test, pred))

print("Mean Squared Error :", mean_squared_error(y_test, pred))
```

The Linear Regression model was trained to predict the number of Rings using the physical measurements of abalones.

The R² Score indicates how much of the variation in the Rings values is explained by the model. A higher R² score means better predictive performance.
The Mean Squared Error (MSE) measures the average squared difference between the actual and predicted values. Lower values indicate better accuracy.


```python
# ==========================================================
# STEP 26: FEATURE ENGINEERING
# Creating a New Feature: Age_Class
# ==========================================================
#
# Feature Engineering is the process of creating
# new features (columns) from existing data.
#
# Here, we convert the numerical 'Rings' column
# into a categorical column called 'Age_Class'.
#
# Rule:
#
# Rings < 10  -> Young
# Rings >= 10 -> Old
#
# This makes the dataset easier to understand
# and can also be useful for classification models.
#
# ==========================================================

# Create a new column called Age_Class

df["Age_Class"] = df["Rings"].apply(
    lambda x: "Young" if x < 10 else "Old"
)

# Display the first few rows
df.head()
```

```python
# ==========================================================
# STEP 27: DECISION TREE CLASSIFICATION
# Predicting Age Class (Young / Old)
# ==========================================================
#
# A Decision Tree is a supervised machine learning
# algorithm used for classification and regression.
#
# Here, we are performing a Classification task.
#
# Target Variable (y):
# Age_Class (Young / Old)
#
# Input Features (X):
# Length
# Diameter
# Height
# Whole_weight
# Shucked_weight
# Viscera_weight
# Shell_weight
#
# ==========================================================

# Import required libraries
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# ----------------------------------------------------------
# Step 1: Select Input Features
# ----------------------------------------------------------

# Remove target column and unnecessary columns
X = df.drop(["Age_Class", "Rings", "Sex"], axis=1)

# Convert categorical variables into numerical values
X = pd.get_dummies(X)

# ----------------------------------------------------------
# Step 2: Select Target Variable
# ----------------------------------------------------------

# Target variable
y = df["Age_Class"]

# ----------------------------------------------------------
# Step 3: Split the Dataset
# ----------------------------------------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

# ----------------------------------------------------------
# Step 4: Create Decision Tree Model
# ----------------------------------------------------------

model = DecisionTreeClassifier(
    random_state=42
)

# ----------------------------------------------------------
# Step 5: Train the Model
# ----------------------------------------------------------

model.fit(X_train, y_train)

# ----------------------------------------------------------
# Step 6: Make Predictions
# ----------------------------------------------------------

pred = model.predict(X_test)

# ----------------------------------------------------------
# Step 7: Evaluate the Model
# ----------------------------------------------------------

accuracy = accuracy_score(y_test, pred)

print("Accuracy :", accuracy)
```
